```python
import pandas as pd



df_gu_facility_counts = pd.read_csv("/home/seo/dev_ws/eda/data/df_gu_facility_counts.csv")
df_car_accident = pd.read_csv("/home/seo/dev_ws/eda/data/df_car_accident.csv")
crime_gu = pd.read_csv("/home/seo/dev_ws/eda/data/crime_gu.csv")
```

```python
df_car_filter = df_car_accident[['구', '발생건수', '정규화', 'rank','교통사고 비율%','인구대비 교통사고 비율_rank']]
df_car_filter
```

```python
df_facility_filter = df_gu_facility_counts[['구', '유흥시설 총합', '정규화', '유흥시설 총합_rank', '면적대비 유흥시설', '면적대비_rank']]
df_facility_filter
```

```python
crime_gu_filter = crime_gu[['구', '발생 합계' , '정규화', '면적대비 범죄', '발생대비_rank', '면적대비_rank']]
crime_gu_filter
```

```python
# '구'를 기준으로 병합
df_merged = df_car_filter[['구', 'rank']].merge(
    df_facility_filter[['구', '유흥시설 총합_rank']], on='구', suffixes=('_car', '_facility')
).merge(
    crime_gu_filter[['구', '발생대비_rank']].rename(columns={'발생대비_rank': 'rank_crime'}), on='구'
)

df_merged.rename(columns={'rank': 'rank_crime'}, inplace=True)

df_merged['rank_avg'] = df_merged[['rank_car', 'rank_facility', 'rank_crime']].mean(axis=1).round(1)

df_merged = df_merged.sort_values(by='rank_avg', ascending=False)

df_merged['total_rank'] = range(1, len(df_merged) + 1)

df_merged[['구', 'rank_avg','total_rank']]
```

교통사고의 발생 건수, 유흥업소의 갯수, 범죄 발생 건수 데이터로 평균을 계산하고 rank로 정렬했을때의 folium 맵

```python
import json
import folium

geo_path = "/home/seo/dev_ws/eda/data/02. skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding="utf-8"))

my_map = folium.Map(
    location=[37.5502, 126.982], zoom_start=11, tiles="OpenStreetMap")

folium.Choropleth(
    geo_data=geo_str,  # GeoJSON 데이터
    data=df_merged,  # 시각화할 데이터
    columns=['구', 'total_rank'],  # 컬럼 지정: 구와 정규화된 유흥시설 수
    key_on="feature.properties.name",  # GeoJSON에서 구 이름에 맞춰 데이터 연결
    fill_color="PuRd", 
    fill_opacity=0.7,  # 색상의 투명도
    line_opacity=0.2,  # 경계선 투명도
    legend_name="total_rank",
).add_to(my_map)

my_map
```

총 인구수 대비 교통사고 비율, 총 면적대비 유흥업소 갯수, 총 면적 대비 범죄횟수 시각화

```python
# '구'를 기준으로 병합
df_merged_option = df_car_filter[['구', '인구대비 교통사고 비율_rank']].merge(
    df_facility_filter[['구', '면적대비_rank']], on='구', suffixes=('_car', '_facility')
).merge(
    crime_gu_filter[['구', '면적대비_rank']].rename(columns={'면적대비_rank': 'rank_crime'}), on='구'
)

# 평균 순위 계산
df_merged_option['rank_avg'] = df_merged_option[['인구대비 교통사고 비율_rank', '면적대비_rank', 'rank_crime']].mean(axis=1).round(1)

# 정렬 후 순위
df_merged_option = df_merged_option.sort_values(by='rank_avg', ascending=False)

# 전체 순위 추가
df_merged_option['total_rank'] = range(1, len(df_merged) + 1)

# 결과 출력
df_merged_option[['구', 'rank_avg', 'total_rank']]
```
