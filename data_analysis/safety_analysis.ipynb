```python
import pandas as pd



df_gu_facility_counts = pd.read_csv("/home/seo/dev_ws/eda/data/df_gu_facility_counts.csv")
df_car_accident = pd.read_csv("/home/seo/dev_ws/eda/data/df_car_accident.csv")
crime_gu = pd.read_csv("/home/seo/dev_ws/eda/data/crime_gu.csv")
```

```python
df_car_filter = df_car_accident[['구', '발생건수', '정규화', 'rank','교통사고 비율%','인구대비 교통사고 비율_rank']]
df_car_filter
```

```python
df_facility_filter = df_gu_facility_counts[['구', '유흥시설 총합', '정규화', '유흥시설 총합_rank', '면적대비 유흥시설', '면적대비_rank']]
df_facility_filter
```

```python
crime_gu_filter = crime_gu[['구', '발생 합계' , '정규화', '면적대비 범죄', '발생대비_rank', '면적대비_rank']]
crime_gu_filter
```

```python
# '구'를 기준으로 병합
df_merged = df_car_filter[['구', 'rank']].merge(
    df_facility_filter[['구', '유흥시설 총합_rank']].rename(columns={'유흥시설 총합_rank': 'rank_facility'}), on='구',
).merge(
    crime_gu_filter[['구', '발생대비_rank']].rename(columns={'발생대비_rank': 'rank_crime'}), on='구'
)

df_merged['rank_avg'] = df_merged[['rank', 'rank_facility', 'rank_crime']].mean(axis=1).round(1)

df_merged = df_merged.sort_values(by='rank_avg', ascending=False)

df_merged['total_rank'] = range(1, len(df_merged) + 1)

df_merged[['구', 'rank_avg','total_rank']]
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns

plt.figure(figsize=(12, 6))
plt.plot(df_merged["구"], df_merged["rank_avg"], marker="o", linestyle="-", color="b", markersize=8)

plt.xticks(rotation=45)
plt.xlabel("구")
plt.ylabel("Rank Average")
plt.title("구별 Rank 평균 변화")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

교통사고의 발생 건수, 유흥업소의 갯수, 범죄 발생 건수 데이터로 평균을 계산하고 rank로 정렬했을때의 folium 맵

```python
import json
import folium

geo_path = "/home/seo/dev_ws/eda/data/02. skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding="utf-8"))

my_map = folium.Map(
    location=[37.5502, 126.982], zoom_start=11, tiles="OpenStreetMap")

folium.Choropleth(
    geo_data=geo_str,  # GeoJSON 데이터
    data=df_merged,  # 시각화할 데이터
    columns=['구', 'total_rank'],  # 컬럼 지정: 구와 정규화된 유흥시설 수
    key_on="feature.properties.name",  # GeoJSON에서 구 이름에 맞춰 데이터 연결
    fill_color="PuRd", 
    fill_opacity=0.7,  # 색상의 투명도
    line_opacity=0.2,  # 경계선 투명도
    legend_name="건수로만 판단한 안전지표 자치구 순위",
).add_to(my_map)

my_map
```

총 인구수 대비 교통사고 비율, 총 면적대비 유흥업소 갯수, 총 면적 대비 범죄횟수 시각화

```python
# '구'를 기준으로 병합
df_merged_option = df_car_filter[['구', '인구대비 교통사고 비율_rank']].merge(
    df_facility_filter[['구', '면적대비_rank']], on='구', suffixes=('_car', '_facility')
).merge(
    crime_gu_filter[['구', '면적대비_rank']].rename(columns={'면적대비_rank': 'rank_crime'}), on='구'
)

# 평균 순위 계산
df_merged_option['rank_avg'] = df_merged_option[['인구대비 교통사고 비율_rank', '면적대비_rank', 'rank_crime']].mean(axis=1).round(1)

# 정렬 후 순위
df_merged_option = df_merged_option.sort_values(by='rank_avg', ascending=False)

# 전체 순위 추가
df_merged_option['total_rank'] = range(1, len(df_merged) + 1)

# 결과 출력
df_merged_option[['구', 'rank_avg', 'total_rank']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# df_merged로 데이터 병합 (이미 병합된 데이터프레임)
df_compare = df_merged[["구", "total_rank"]].merge(
    df_merged_option[["구", "total_rank"]], on="구", suffixes=("_original", "_option")
)

# 정렬 (total_rank_original 기준, 높은 순위가 오른쪽에 오도록)
df_compare = df_compare.sort_values(by="total_rank_original", ascending=False)

fig, ax1 = plt.subplots(figsize=(14, 7))

# total_rank_original에 맞춰 막대그래프 그리기
sns.barplot(x=df_compare["구"], y=df_compare["total_rank_original"], ax=ax1, color="skyblue", label="Original Rank")

ax1.set_xlabel("구")
ax1.set_ylabel("구별 순위_기본값")
ax1.set_title("구별 Total Rank 막대: 구별 순위, 꺾은선: 구별순위_변수 적용)")

ax1.set_xticklabels(df_compare["구"], rotation=45)

# 각 막대 위에 숫자 추가
for i, row in enumerate(df_compare.itertuples()):
    ax1.text(i, row.total_rank_original + 0.1, str(row.total_rank_original), ha='center', va='bottom', fontsize=12)

# 꺾은선 그래프 (Option Rank)
ax2 = ax1.twinx()
ax2.plot(df_compare["구"], df_compare["total_rank_option"], marker="o", linestyle="--", color="red", label="Option Rank")

ax2.set_ylabel("구별 순위_변수 적용")

# 꺾은선 그래프에 순위 텍스트 추가 (y값을 조금 올려서 겹치지 않도록)
for i, row in enumerate(df_compare.itertuples()):
    ax2.text(i, row.total_rank_option + 0.2, str(row.total_rank_option), ha='center', va='bottom', fontsize=12, color="red")

# 범례 추가
ax1.legend(loc="upper left")  # 왼쪽 상단에 배치
ax2.legend(loc="upper right")  # 오른쪽 상단에 배치

plt.savefig('rank_comparison_graph.png', bbox_inches='tight')

plt.show()
```
