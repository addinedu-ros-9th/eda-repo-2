```python
import pandas as pd

df_gu_facility_counts = pd.read_csv("/home/seo/dev_ws/eda/data/df_gu_facility_counts.csv")
df_car_accident = pd.read_csv("/home/seo/dev_ws/eda/data/df_car_accident.csv")
crime_gu = pd.read_csv("/home/seo/dev_ws/eda/data/crime_gu.csv")
```

```python
df_car_filter = df_car_accident[['구', '발생건수', '정규화', 'rank','교통사고 비율%','인구대비 교통사고 비율_rank']]
df_car_filter
```

```python
df_gu_facility_counts = df_gu_facility_counts.rename(columns={"gu_id": "구", "rank": "유흥시설 총합_rank"})
df_facility_filter = df_gu_facility_counts[['구', '유흥시설 총합', '정규화', '유흥시설 총합_rank', '면적대비 유흥시설', '면적대비_rank']]

df_facility_filter
```

```python
crime_gu_filter = crime_gu[['구', '발생 합계' , '정규화', '면적대비 범죄', '발생대비_rank', '면적대비_rank']]
crime_gu_filter
```

```python
# '구'를 기준으로 병합
df_merged = df_car_filter[['구', 'rank']].merge(
    df_facility_filter[['구', '유흥시설 총합_rank']].rename(columns={'유흥시설 총합_rank': 'rank_facility'}), on='구',
).merge(
    crime_gu_filter[['구', '발생대비_rank']].rename(columns={'발생대비_rank': 'rank_crime'}), on='구'
)

df_merged['rank_avg'] = df_merged[['rank', 'rank_facility', 'rank_crime']].mean(axis=1).round(1)

df_merged = df_merged.sort_values(by='rank_avg', ascending=False)

df_merged['total_rank'] = range(1, len(df_merged) + 1)

df_merged[['구', 'rank_avg','total_rank']]
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns

plt.figure(figsize=(12, 6))
plt.plot(df_merged["구"], df_merged["rank_avg"], marker="o", linestyle="-", color="b", markersize=8)

plt.xticks(rotation=45)
plt.xlabel("구")
plt.ylabel("Rank Average")
plt.title("구별 Rank 평균 변화")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
