```python
df_facility['도로명주소'].replace('', None, inplace=True)  # 빈 문자열을 None으로 변환
df_facility = df_facility.dropna(subset=['도로명주소'])  # NaN 값이 있는 행 삭제
```

```python
import pandas as pd

df_facility['구'] = df_facility['도로명주소'].str.extract(r'(\S+구)')
# '업태구분명'에 포함된 '룸살롱', '간이주점', '클럽or나이트'와 기타의 개수를 구별로 세는 코드
df_gu_facility_counts = df_facility.groupby('구').apply(
    lambda x: pd.Series({
        '룸살롱': (x['업태구분명'].str.contains('룸살롱', na=False)).sum(),
        '간이주점': (x['업태구분명'].str.contains('간이주점', na=False)).sum(),
        '클럽,나이트': (x['업태구분명'].str.contains('클럽|나이트', na=False)).sum(),
        '기타': (x['업태구분명'].str.contains('룸살롱|간이주점|클럽|나이트', na=False) == False).sum()
    })
).reset_index()

df_gu_facility_counts['유흥시설 총합'] = df_gu_facility_counts[['룸살롱', '간이주점', '클럽,나이트', '기타']].sum(axis=1)
#cols = ['구', '유흥시설 총합', '룸살롱', '간이주점', '클럽,나이트', '기타']
#gu_counts = gu_counts[cols]

df_gu_facility_counts
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

df_gu_facility_counts.set_index("구")["유흥시설 총합"].sort_values().plot(kind="barh", grid=True, figsize=(10, 10))

plt.xlabel("유흥시설 개수")  # x축 라벨 설정
plt.ylabel("구")  # y축 라벨 설정 (필요하면 추가)
plt.title("구별 유흥시설 개수")  # 그래프 제목 추가|
plt.show()
```

```python
df_gu_facility_counts.set_index("구")["유흥시설 총합"].sort_values().plot(kind="barh", grid=True, figsize=(10, 10))

plt.xlabel("유흥시설 개수")  # x축 라벨 설정
plt.ylabel("구")  # y축 라벨 설정 (필요하면 추가)
plt.title("구별 유흥시설 개수")  # 그래프 제목 추가|
plt.show()  
```

```python
# 전체 유흥시설 총합 구하기
total_facilities = df_gu_facility_counts['유흥시설 총합'].sum()

# 각 구별 유흥시설 총합의 비율 계산
df_gu_facility_counts['비율'] = df_gu_facility_counts['유흥시설 총합'] / total_facilities * 100

# 구별 유흥시설 개수의 비율을 내림차순으로 정렬하여 그래프를 그립니다.
df_gu_facility_counts.set_index("구")["비율"].sort_values().plot(kind="barh", grid=True, figsize=(10, 10))

# x축 라벨과 y축 라벨 설정
plt.xlabel("유흥시설 비율 (%)")  # x축 라벨을 비율로 설정
plt.ylabel("구")  # y축 라벨을 '구'로 설정
plt.title("구별 유흥시설 비율")  # 그래프 제목 설정

# 그래프 출력
plt.show()
```

```python
import json
import folium


geo_path = "./data/02. skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding="utf-8"))

my_map = folium.Map(
    location=[37.5502, 126.982], zoom_start=11, tiles="OpenStreetMap")

folium.Choropleth(
    geo_data=geo_str,  # GeoJSON 데이터
    data=df_gu_facility_counts,  # 시각화할 데이터
    columns=['구', '유흥시설 총합'],  # 컬럼 지정: 구와 정규화된 유흥시설 수
    key_on="feature.properties.name",  # GeoJSON에서 구 이름에 맞춰 데이터 연결
    fill_color="PuRd",
    fill_opacity=0.7,  # 색상의 투명도
    line_opacity=0.2,  # 경계선 투명도
    legend_name="유흥시설 총합",  # 범례 이름
).add_to(my_map)

my_map
```
