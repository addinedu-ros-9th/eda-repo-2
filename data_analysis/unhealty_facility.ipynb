```python
df_facility = df_facility[df_facility['상세영업상태명'].str.contains('영업')]
df_facility.head()
```

```python
df_facility = df_facility.drop(columns=['개방자치단체코드', '관리번호', "인허가일자", "인허가취소일자", "영업상태코드",
                       "영업상태명", "상세영업상태코드", #"상세영업상태명", 
                       "폐업일자","휴업시작일자","휴업종료일자","재개업일자", "전화번호",
                         "소재지면적", "소재지우편번호", "지번주소","도로명우편번호", "최종수정일자",
                           "데이터갱신구분", "데이터갱신일자", "좌표정보(X)", "좌표정보(Y)", "남성종사자수",
                             "여성종사자수", "등급구분명", "급수시설구분명","총인원", "본사종업원수",
                               "공장사무직종업원수", "공장판매직종업원수", "공장생산직종업원수","건물소유구분명", "보증액",
                                 "월세액", "다중이용업소여부", "시설총규모", "전통업소지정번호", "전통업소주된음식", "홈페이지" ])
df_facility
```

```python
df_facility['도로명주소'].replace('', None, inplace=True)  # 빈 문자열을 None으로 변환
df_facility = df_facility.dropna(subset=['도로명주소'])  # NaN 값이 있는 행 삭제
```

```python
import pandas as pd

df_facility['구'] = df_facility['도로명주소'].str.extract(r'(\S+구)')
# '업태구분명'에 포함된 '룸살롱', '간이주점', '클럽or나이트'와 기타의 개수를 구별로 세는 코드
df_gu_facility_counts = df_facility.groupby('구').apply(
    lambda x: pd.Series({
        '룸살롱': (x['업태구분명'].str.contains('룸살롱', na=False)).sum(),
        '간이주점': (x['업태구분명'].str.contains('간이주점', na=False)).sum(),
        '클럽,나이트': (x['업태구분명'].str.contains('클럽|나이트', na=False)).sum(),
        '기타': (x['업태구분명'].str.contains('룸살롱|간이주점|클럽|나이트', na=False) == False).sum()
    })
).reset_index()

df_gu_facility_counts['유흥시설 총합'] = df_gu_facility_counts[['룸살롱', '간이주점', '클럽,나이트', '기타']].sum(axis=1)
#cols = ['구', '유흥시설 총합', '룸살롱', '간이주점', '클럽,나이트', '기타']
#gu_counts = gu_counts[cols]

df_gu_facility_counts["정규화"] = df_gu_facility_counts["유흥시설 총합"] / df_gu_facility_counts["유흥시설 총합"].max()

df_gu_facility_counts
```

```python
df_gu_area = pd.read_csv("/home/seo/dev_ws/eda/data/gu_area.csv")
df_gu_area = df_gu_area.drop(columns=['자치구별(1)', '2023.1', "2023.2", "2023.3", "2023.4","2023.5"])
df_gu_area = df_gu_area.drop(index = [0, 1, 2])
df_gu_area = df_gu_area.reset_index(drop=True)
df_gu_area
```

```python
df_gu_area = df_gu_area.rename(columns= {"자치구별(2)":"구", "2023": "면적"})
df_gu_area
```

```python
df_gu_area["면적"] = pd.to_numeric(df_gu_area["면적"], errors='coerce')

df_gu_facility_counts["면적대비 유흥시설"]=df_gu_facility_counts["유흥시설 총합"] / df_gu_area["면적"]

df_gu_facility_counts["rank"] = df_gu_facility_counts['유흥시설 총합'].rank(ascending=False, method='min')

df_gu_facility_counts["면적대비_rank"] = df_gu_facility_counts['면적대비 유흥시설'].rank(ascending=False, method='min')

#df_gu_facility_counts.rename(columns={"구": "gu_id"}, inplace=True)
df_gu_facility_counts
```
