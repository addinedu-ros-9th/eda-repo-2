```python
import pandas as pd

df_car_accident = pd.read_csv("/home/seo/dev_ws/eda/data/car_accident.csv")

df_car_accident = df_car_accident.drop(columns=['자치구별(1)', '2023.1', "2023.3", "2023.5"])

df_car_accident.rename(columns={"자치구별(2)": "구", "2023": "발생건수", "2023.2": "사망자수", "2023.4": "부상자수"}, inplace=True)

df_car_accident.drop(index=[0, 1], inplace=True)

df_car_accident.reset_index(drop=True, inplace=True)

# "발생건수"를 숫자로 변환
df_car_accident["발생건수"] = pd.to_numeric(df_car_accident["발생건수"], errors="coerce")

total_accidents = df_car_accident["발생건수"].sum()

# 각 구의 교통사고 비율 구하기
df_car_accident["교통사고비율"] = round((df_car_accident["발생건수"] / total_accidents) * 100, 2)

df_car_accident['정규화'] = df_car_accident['발생건수'] / df_car_accident['발생건수'].max()

df_car_accident['rank'] = df_car_accident['정규화'].rank(ascending=False)

df_car_accident['rank'] = df_car_accident['rank'].astype(int)

df_car_accident

df_car_accident.to_csv("/home/seo/dev_ws/eda/data/df_car_accident.csv", index=False, encoding ='utf-8')
```

```python
df_population = pd.read_csv("/home/seo/dev_ws/eda/data/population.csv", encoding = 'euc-kr')

df_population['기준일ID'] = df_population['기준일ID'].astype(str)
df_population['시간대구분'] = df_population['시간대구분'].astype(str)

df_population = df_population[df_population['기준일ID'].str.contains('20250321')]
df_population = df_population[df_population['시간대구분'].str.contains('^0$')]

df_population['자치구코드'] = df_population['자치구코드'].astype(str)

gu_code_mapping = {
    "11110": "종로구",
    "11140": "중구",
    "11170": "용산구",
    "11200": "성동구",
    "11215": "광진구",
    "11230": "동대문구",
    "11260": "중랑구",
    "11290": "성북구",
    "11305": "강북구",
    "11320": "도봉구",
    "11350": "노원구",
    "11380": "은평구",
    "11410": "서대문구",
    "11440": "마포구",
    "11470": "구로구",
    "11500": "동작구",
    "11530": "관악구",
    "11545": "서초구",
    "11560": "강남구",
    "11590": "송파구",
    "11620": "강동구",
    "11650": "광진구",
    "11680": "중랑구",
    "11710": "성북구",
    "11740": "강북구"
}

# 자치구 코드가 있는 칼럼을 변환
df_population['구'] = df_population['자치구코드'].map(gu_code_mapping)

df_population[['자치구코드','구', '총생활인구수']]
```

```python
# '자치구' 칼럼을 기준으로 두 데이터프레임 병합
merged_df = pd.merge(df_car_accident, df_population[['구', '총생활인구수']], on='구', how='inner')

# 구별로 교통사고 비율 계산: 발생건수 / 총생활인구수
merged_df['교통사고 비율%'] = merged_df['발생건수'] / merged_df['총생활인구수']

# 교통사고 비율을 백분율로 변환 (필요시)
merged_df['교통사고 비율%'] = merged_df['교통사고 비율%'] * 100

merged_df['교통사고 비율%'] = merged_df['교통사고 비율%'].round(2)

merged_df['인구대비 교통사고 비율_rank'] = merged_df['교통사고 비율%'].rank(ascending=False)

merged_df['rank'] = merged_df['rank'].astype(int)

merged_df
```

```python
merged_df.to_csv("/home/seo/dev_ws/eda/data/df_car_accident.csv", index=False, encoding ='utf-8')
```
