```python
def insert_dong(gu, dong):
    # 구 ID 가져오기
    cur.execute('SELECT gu_id FROM gu WHERE gu_name = %s', (gu,))
    gu_id = cur.fetchone()[0]

    # 도로명 테이블에 도로명과 구 ID 삽입
    cur.execute('INSERT INTO dong (gu_id, dong_name) VALUES (%s, %s)', (gu_id, dong))
    conn.commit()
```

```python
for gu, dong in dong_list:
    insert_dong(gu, dong)
```
