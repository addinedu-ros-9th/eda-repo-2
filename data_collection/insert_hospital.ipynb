```python
def insert_hospital(hospital_name, hospital_type, gu_name, dong_name, address):
    check=[]

    # 구 ID 가져오기
    cur.execute('SELECT gu_id FROM gu WHERE gu_name = %s', (gu_name,))
    gu_id = cur.fetchone()[0]

    #동 ID 가져오기
    cur.execute('SELECT dong_id FROM dong WHERE dong_name = %s', (dong_name,))
    dong_result = cur.fetchone()

    if dong_result is None:
        print(f"동명 {dong_name}이(가) 데이터베이스에 없습니다.")
        check.append(0)
    else:    
        dong_id = dong_result[0]
        check.append(1)


    doro_name = address.split(" ")[2]
    #도로 ID 가져오기
    cur.execute('SELECT doro_id FROM doro WHERE doro_name = %s', (address.split(" ")[2],))
    # doro_id = cur.fetchone()[0]
    doro_result = cur.fetchone()
    
    if doro_result is None:
        print(f"도로명 {doro_name}이(가) 데이터베이스에 없습니다.")
        check.append(0)
    else:    
        doro_id = doro_result[0]
        check.append(1)
    
    if check[0]==0:
        cur.execute(
                'INSERT INTO hospital (hospital_type, hospital_name, hospital_address, gu_id, doro_id) '
                'VALUES (%s, %s, %s, %s, %s)', 
                (hospital_type, hospital_name, address, gu_id, doro_id))
    elif check[1]==0:
        cur.execute(
                'INSERT INTO hospital (hospital_type, hospital_name, hospital_address, gu_id, dong_id) '
                'VALUES (%s, %s, %s, %s, %s)', 
                (hospital_type, hospital_name, address, gu_id, dong_id))
    elif check[0]==0 and check[1]==0:
        cur.execute(
                'INSERT INTO hospital (hospital_type, hospital_name, hospital_address, gu_id) '
                'VALUES (%s, %s, %s, %s)', 
                (hospital_type, hospital_name, address, gu_id))
    else:
        cur.execute(
                'INSERT INTO hospital (hospital_type, hospital_name, hospital_address, gu_id, doro_id, dong_id) '
                'VALUES (%s, %s, %s, %s, %s, %s)', 
                (hospital_type, hospital_name, address, gu_id, doro_id, dong_id))
    
    conn.commit()
```

```python
for hospital_name, hospital_type, gu_name, dong_name, address in hopital_list:
    insert_hospital(hospital_name, hospital_type, gu_name, dong_name, address)
```
