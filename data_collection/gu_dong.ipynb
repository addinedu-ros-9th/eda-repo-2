```python
from tqdm import tqdm

def insert_gu(gu_list):
    # 구 데이터 삽입 (중복 방지)
    values = [(gu_name,) for gu_name in gu_list]  # gu_name을 튜플로 만들어 리스트로 변환
    
    # 한 번에 여러 값 삽입
    cur.executemany('INSERT INTO gu (gu_name) VALUES (%s)', values)
    
    # 한 번에 commit
    conn.commit()

# 구 데이터 삽입
insert_gu(gu_list)

```

```python
import pandas as pd

file_path = "../raw_data/rnaddrkor_seoul.txt"

# 파일을 읽고, '|' 구분자로 나누기
df = pd.read_csv(file_path, sep='|', encoding='cp949', header=None)

# 자치구(4번째 컬럼)와 도로명(11번째 컬럼)만 추출하기
df_selected = df[[3, 10]]

# 컬럼 이름을 지정
df_selected.columns = ['gu_name', 'road_name']

df_selected
```

```python
doro = df_selected[['gu_name', 'road_name']].drop_duplicates().reset_index(drop=True)
```

```python
doro
```

```python
doro_list = [(row['gu_name'], row['road_name']) for _, row in doro.iterrows()]
```

```python
doro_list
```

```python
def insert_doro(gu_name, doro):
    # 구 ID 가져오기
    cur.execute('SELECT gu_id FROM gu WHERE gu_name = %s', (gu_name,))
    gu_id = cur.fetchone()[0]

    # 도로명 테이블에 도로명과 구 ID 삽입
    cur.execute('INSERT INTO doro (gu_id, doro_name) VALUES (%s, %s)', (gu_id, doro))
    conn.commit()
```

```python
for gu_name, road_name in doro_list:
    insert_doro(gu_name, road_name)
```
