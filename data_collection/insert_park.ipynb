```python
import pandas as pd

df = pd.read_excel('../../data/park.xlsx', skiprows=3)

df.head()
```

```python
df.rename(columns={df.columns[0]: "gu_name"}, inplace=True)


df.columns = ["gu_name", "공원_면적", "공원_1인당면적", 
              "도시공원_면적", "도시공원_1인당면적", 
              "도보생활권공원_면적", "도보생활권공원_1인당면적"]

df.reset_index(drop=True, inplace=True)

print(df.head())
```

melt()를 사용해 기존 컬럼을 park_type, value로 변환

```python
df_melted = df.melt(id_vars=["gu_name"], 
                    var_name="park_type", 
                    value_name="value")

df_melted["park_size"] = df_melted["value"].where(df_melted["park_type"].str.contains("면적"))
df_melted["park_for_one"] = df_melted["value"].where(df_melted["park_type"].str.contains("1인당면적"))

df_melted["park_type"] = df_melted["park_type"].str.replace("_면적", "").str.replace("_1인당면적", "")

df_melted.drop(columns=["value"], inplace=True)
df_melted.dropna(subset=["park_size", "park_for_one"], how="all", inplace=True)
df_melted.reset_index(drop=True, inplace=True)
```

```python
df_melted
```

```python
df_long = df.melt(id_vars=["gu_name"], var_name="category", value_name="value")

# park_type, park_size, park_for_one 분리
df_long["park_type"] = df_long["category"].str.extract(r"(.*)_")[0]
df_long["park_for_one"] = df_long["category"].str.contains("1인당면적")

# 면적과 1인당 면적을 나누어 저장
df_wide = df_long.pivot_table(index=["gu_name", "park_type"], 
                              columns="park_for_one", values="value", 
                              aggfunc="first").reset_index()

# 컬럼명 변경
df_wide.columns = ["gu_name", "park_type", "park_size", "park_for_one"]

# 데이터 정리
df_wide = df_wide.sort_values(["gu_name", "park_type"]).reset_index(drop=True)

df_wide
```

```python
def insert_park(gu, park_type, park_size, park_for_one):
    # 구 ID 가져오기
    cur.execute('SELECT gu_id FROM gu WHERE gu_name = %s', (gu,))
    gu_id = cur.fetchone()[0]

    cur.execute(
            'INSERT INTO park (gu_id, park_type, park_size, park_for_one)'
            'VALUES (%s, %s, %s, %s)', 
            (gu_id, park_type, park_size, park_for_one))
    
    conn.commit()
```
